## 다중 입력

- 독립 변수 x가 2개 이상인 경우에 대한 학습

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

### 다중 선형 회귀

- y를 결정하는 데 있어 독립 변수가 3개인 선형 회귀 풀기



- 3개의 특성을 가진 벡터 [x1, x2, x3]를 대문자 X로 표기.

```python
# 중간 고사, 기말 고사, 가산점 점수

X = np.array([[70, 85, 11], [71, 89, 18], [50, 80, 20], [99, 20, 10], [50, 10, 10]])
y = np.array([73, 82, 72, 57, 34]) # 최종 성적

model = Sequential()
model.add(Dense(1, input_dim = 3, activation = 'linear'))

sgd = optimizers.SGD(lr = 0.0001)
model.compile(optimizer = sgd, loss = 'mse', metrics = ['mse'])
model.fit(X, y, epochs = 2000)
```

```python
print(model.predict(X))
```

- 실제값에 근접한 예측을 하는 것을 볼 수 있음. 훈련할 때 사용하지 않았던 데이터를 가지고 예측 작업을 수행.

```python
X_test = np.array([[20, 99 ,10], [40, 50, 20]])
print(model.predict(X_test))
```

## 다중 로지스틱 회귀

- y를 결정하는 데 있어 독립 변수 x가 2개인 로지스틱 회귀 풀기
- 꽃받침(Sepal)의 길이와 꽃잎(Petal)의 길이와 해당 꽃이 A인지 B인지가 적혀져 있는 데이터가 있을 때, 새로 조사한 꽃받침의 길이와 꽃잎의 길이로부터 무슨 꽃인지 예측하는 모델 생성



- 2개의 특성을 가진 벡터 [x1, x2]를 대문자 X로 표기

```python
# 독립 변수가 2개인 새로운 예를 들어 케라스로 구현
# 두 개의 입력 x1, x2의 합이 2 이상이면 출력값 y가 1이 되고 두 개의 입력의 합이 2 미만인 경우에만 출력값이 0이 되는 로직을 구현

X = np.array([[0, 0], [0, 1], [1, 0], [0, 2], [1, 1], [2, 0]])
y = np.array([0, 0, 0, 1, 1, 1])

model = Sequential()
model.add(Dense(1, input_dim = 2, activation = 'sigmoid'))
model.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])

model.fit(X, y, epochs = 2000)
```

```python
print(model.predict(X))
```

- 입력의 합이 2 이상인 경우에는 전부 값이 0.5를 넘는 것을 볼 수 있음.
